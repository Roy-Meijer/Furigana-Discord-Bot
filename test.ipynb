```python
import discord
from discord.ext import commands
import pykakasi
import re

kks = pykakasi.kakasi()

# Check if word has Kanji
def contains_kanji(word):
    return re.search(r'[\u4E00-\u9FBF]', word)

# Generate Furigana
def get_furigana_text(sentence):
    word_list = kks.convert(sentence)
    formatted_text = "".join([
        f"{item['orig']}({item['hira']})" if contains_kanji(item['orig']) else item['orig']
        for item in word_list
    ])

    print(formatted_text)
    return formatted_text
```
